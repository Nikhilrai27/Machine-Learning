```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
##creating the dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=25)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred=logistic.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

#Hyperparameter tunning and cross validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,.01]
solver=['newton-cg','lbfgs' ,'liblinear' 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
#gridsearch cv
grid=GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=cv,n_jobs=-1)
```

```python
print(grid)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_=grid.predict(x_test)
```

```python
score=accuracy_score(y_test,y_pred_)
print(score)
cm=confusion_matrix(y_test,y_pred_)
print(cm)
print(classification_report(y_test,y_pred_))
```

#Randomized search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(x_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred__=randomcv.predict(x_test)
```

```python
score=accuracy_score(y_test,y_pred__)
print(score)
cm=confusion_matrix(y_test,y_pred__)
print(cm)
print(classification_report(y_test,y_pred__))
```
